### Visualización de datos

```python
x = df.drop(["target"],axis=1) #x are the other 30 columns different that target
y = df["target"] #y is the target column
```

```python
#Graph target values
sns.countplot(data=df, x='target', palette="Blues") #Show the counts of observations in each categorical bin using bars.
plt.xlabel('Target', fontsize=14, weight='bold') #Parameters of x axis
plt.ylabel('Count', fontsize=14, weight='bold') #Parameters of y axis
plt.xticks(ticks=[0, 1], labels=['No Cancer', 'Cancer'], weight='bold') #Set values 0 and 1 to labels No Cancer and Cancer, respectively

# add labels on each bar
for bar in plt.gca().containers:
    label = [f"{int(h.get_height())}" for h in bar] #label: string literal of height of bar
    plt.bar_label(bar, labels=label, label_type='edge') #set plot bar parameters
plt.show() #visualize the plot
```

```python
## We define the list of numerical characteristics
# We define the size of the figure
plt.figure(figsize=(24, 12))

# We add the title
plt.suptitle('Comparison of distributions of numerical characteristics by target value', fontsize=14)

# We create a loop to create the sub-graphs
for i,feature in enumerate(x.columns):
    ax = plt.subplot(5, 6, i+1)
    sns.boxplot(x='target', y=feature, data=df, palette="Blues", ax=ax)
    ax.set_xlabel('Target', fontsize=12)
    ax.set_ylabel(feature, fontsize=12)

# We adjust the design and show the figure
plt.tight_layout()
plt.show()
```

```python
## Checking numerical features distribution
# We define the size of the figure
plt.figure(figsize = (18, 9))
# We define a title
plt.suptitle('Distribution of numerical characteristics', fontsize=16)

# We create a loop to create the sub-graphs
for i,feature in enumerate(x.columns):
  ax = plt.subplot(5, 6, i+1)
  sns.histplot(df[feature], color='blue', kde=True, ax=ax) # data=num_ft[i] -> values of the columns excluding column target
  ax.set_title(f"skew={df[feature].skew():.2f}", fontsize=10) #calculates the skew for each column and represent it literal
  ax.set_xlabel(feature, fontsize=8) #name of column
  ax.set_ylabel('Frequency', fontsize=8)
    
# We adjust the design and show the figure
plt.tight_layout()
plt.show()

#  it is observed that most of the predictor variables are normally distributed (bell shape)
```

```python
#Correlation between variables
corr = df.corr()# we calculate the correlation of the data
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='coolwarm')# we generate the plot
plt.title('Correlation Matrix')
plt.show()
```

```python
#Normalization of data
normalized_x = (x-x.min())/(x.max()-x.min()) 
normalized_x
```

### LazyPredict model test

```python
!pip -q install lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(normalized_x, y,test_size=.3,random_state =123)
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

### Stochastic Gradient Descent Classifier model test

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve

sgd = SGDClassifier(loss='log', random_state=42)
sgd.fit(X_train, y_train)
y_pred = sgd.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

y_proba = sgd.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_proba)
print("ROC AUC:", roc_auc)

fpr, tpr, thresholds = roc_curve(y_test, y_proba)
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```
